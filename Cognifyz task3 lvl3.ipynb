# Step 1: Import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Step 2: Load the dataset

```python

dataset = pd.read_csv("C:\\Users\\Narthana\\Downloads\\Dataset.csv")
```

# Step 3: Explore the dataset

```python
print(dataset.head())
```

# Step 4: Data cleaning and preprocessing

```python
# Data cleaning and preprocessing
dataset.dropna(inplace=True)
```

```python
# Convert categorical variables to numerical
dataset['Has Online delivery'] = dataset['Has Online delivery'].map({'Yes': 1, 'No': 0})
dataset['Has Table booking'] = dataset['Has Table booking'].map({'Yes': 1, 'No': 0})
```

# Step 5: Data analysis

```python
# Data analysis
analysis_df = dataset.groupby('Price range')[['Has Online delivery', 'Has Table booking']].mean().reset_index()
```

```python
# Print the result
print(analysis_df)
```

# Step 6: Visualization

```python
# Visualization
# Bar plot for Online Delivery
sns.barplot(x='Price range', y='Has Online delivery', data=analysis_df)
plt.title('Online Delivery Availability vs. Price Range')
plt.show()

# Bar plot for Table Booking
sns.barplot(x='Price range', y='Has Table booking', data=analysis_df)
plt.title('Table Booking Availability vs. Price Range')
plt.show()
```

```python
# Convert categorical variables to numerical
dataset['Has Online delivery'] = dataset['Has Online delivery'].map({'Yes': 1, 'No': 0})
dataset['Has Table booking'] = dataset['Has Table booking'].map({'Yes': 1, 'No': 0})

# Data analysis
analysis_df = dataset.groupby('Price range').agg({
    'Has Online delivery': 'mean',
    'Has Table booking': 'mean',
    'Aggregate rating': 'mean',
    'Votes': 'sum'
}).reset_index()

# Print the result
print(analysis_df)

# Visualization

# Bar plot for Online Delivery
sns.barplot(x='Price range', y='Has Online delivery', data=analysis_df)
plt.title('Online Delivery Availability vs. Price Range')
plt.show()

# Bar plot for Table Booking
sns.barplot(x='Price range', y='Has Table booking', data=analysis_df)
plt.title('Table Booking Availability vs. Price Range')
plt.show()

# Bar plot for Average Rating
sns.barplot(x='Price range', y='Aggregate rating', data=analysis_df)
plt.title('Average Rating vs. Price Range')
plt.show()

# Scatter plot for Votes vs. Price Range
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Price range', y='Votes', data=analysis_df, size='Votes', hue='Votes', palette='viridis', sizes=(20, 200))
plt.title('Votes vs. Price Range')
plt.show()

# Correlation heatmap
correlation_matrix = dataset[['Has Online delivery', 'Has Table booking', 'Aggregate rating', 'Votes']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

Online Delivery and Table Booking:

The availability of online delivery and table booking was analyzed based on different price ranges.
Bar plots were created to visualize the relationship between the price range and the availability of online delivery and table booking.

Average Rating:

The average rating for each price range was calculated and visualized using a bar plot.
This provides insights into how the average rating varies across different price ranges.

Votes vs. Price Range:

A scatter plot was generated to explore the relationship between the number of votes and the price range.
The size of the points in the scatter plot represents the total number of votes for each price range.
Correlation Matrix:

A correlation matrix was computed to understand the correlation between variables such as online delivery, table booking, average rating, and votes.
A heatmap was created to visually represent the correlation matrix.
